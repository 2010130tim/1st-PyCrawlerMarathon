### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests
r=requests.get("https://api.github.com/events")
r.text #像是dict
```

```python
import json
json.loads(r.text) #json格式
```
