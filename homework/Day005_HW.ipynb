## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")
r.text

import json
data=json.loads(r.text)
print('這個 API 一次會回傳',len(data),'筆資料'+'\n')
print('每一筆資料包含下列欄位:')
print(data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」


for d in data:
    
    print('標題: '+d['title'])
    print('貼文時間: '+d['createdAt'])
    print('留言人數:',d['commentCount'])
    print('按讚人數:',d['likeCount'])
    print('')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
comment_sum=0
like_sum=0

for d in data:
    comment_sum+=d['commentCount']
    sum=comment_sum
    
print('熱門文章')
print('平均留言人數:', sum/len(data))

for d in data:
    like_sum+=d['likeCount']
    sum=like_sum
print('平均按讚人數:', sum/len(data))
print("")


ra=requests.get("https://www.dcard.tw/_api/forums/pet/posts?popular=false")
data=json.loads(ra.text)

comment_sum=0
like_sum=0
sum=0

for d in data:
    comment_sum+=d['commentCount']
    sum=comment_sum
    
print('非熱門文章')
print('平均留言人數:', sum/len(data))
sum=0

for d in data:
     like_sum+=d['likeCount']
     sum=like_sum
print('平均按讚人數:', sum/len(data))
```
